```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Example: Generate some sample time series data (monthly sales data)
np.random.seed(0)
date_range = pd.date_range(start='2020-01-01', periods=36, freq='M')
sales_data = pd.Series(100 + np.sin(np.arange(36)) * 10 + np.random.normal(scale=5, size=36), index=date_range)

# Plot the original data
plt.figure(figsize=(10, 6))
plt.plot(sales_data, label='Original Sales Data')
plt.title('Monthly Sales Data')
plt.legend()
plt.show()
```

```python
# Decompose the time series using an additive model
decomposition = seasonal_decompose(sales_data, model='additive', period=12)
```

```python
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend', color='green')
plt.legend(loc='upper left')
```

```python
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonality', color='orange')
plt.legend(loc='upper left')
```

```python
plt.subplot(414)
plt.plot(decomposition.resid, label='Residuals', color='red')
plt.legend(loc='upper left')
```
